Senaryo2: Endüstriyel Boya Karışımı
Bir fabrika, iki tür pigment karışımıyla ideal renk yoğunluğunu yakalamak istiyor.

• Amaç fonksiyonu:

y = 5x₁ + 2x₂ - x₁x₂ (renk kalitesi puanı)

• Değişkenler:

o x₁: Pigment A oranı (%) → [0, 100]

o x₂: Pigment B oranı (%) → [0, 100]

• Kısıtlar:

o x₁ + x₂ = 100 (Karışım toplamı %100 olmalı)
o x₁ ≥ 30 (A pigmenti asgari kullanılmalı)

**Problem Tanımı**

Bu çalışmada, iki farklı pigmentin belirli oranlarda karıştırılmasıyla elde edilen boyanın renk kalitesi puanının maksimize edilmesi amaçlanmıştır. Problem, doğrusal olmayan bir amaç fonksiyonuna ve oran kısıtlarına sahip olduğundan, çözüm için Genetik Algoritma (GA) yöntemi tercih edilmiştir.

Amaç, verilen kısıtlar altında pigment oranlarını optimize ederek en yüksek kalite skorunu elde etmektir.

**Amaç Fonksiyonu**

Optimize edilmek istenen amaç fonksiyonu aşağıdaki gibidir:

y = 5x1 + 2x2 - x1x2

Burada:

- x1: Pigment A oranı (%)  
- x2: Pigment B oranı (%)  

Fonksiyondaki ilk iki terim pigmentlerin renk kalitesine olan katkısını temsil ederken,  
(-x1 x2) terimi pigmentlerin birlikte kullanımından kaynaklanan etkileşim maliyetini ifade etmektedir.


**Kısıtlar**

Problem aşağıdaki kısıtlar altında çözülmüştür:

- x1+x2=100 (Toplam karışım oranı)
- x1​≥30(Pigment A için minimum kullanım oranı)
- x2​≥0

Kısıtları ihlal eden çözümler, Genetik Algoritma içerisinde **ceza fonksiyonu** ile düşük uygunluk (fitness) değerine sahip olacak şekilde değerlendirilmiştir.


**Genetik Algoritma Yapısı**

Kullanılan Genetik Algoritma aşağıdaki temel adımlardan oluşmaktadır:

- Başlatma: $[0, 100]$ aralığında rastgele başlangıç popülasyonunun oluşturulması.
- Uygunluk Hesaplama: Amaç fonksiyonu eksi Ceza Fonksiyonu yöntemi ile uygunluk (fitness) hesaplanması.
- Seçilim: Turnuva Seçimi yöntemi ile ebeveynlerin seçilmesi.  
- Çaprazlama: Gerçek değerli GA için uygun olan Ağırlıklı Ortalama (Aritmetik) yöntemi ile yeni bireylerin oluşturulması.
- Mutasyon: Düşük bir olasılıkla Gaussian Mutasyon uygulanması ve genlerin sınır değerlerinde (clip) tutulması.
- Elitizm: Her neslin en iyi bireyinin korunarak doğrudan bir sonraki nesle aktarılması.
- Döngü: Belirlenen Yüksek Jenerasyon Sayısı boyunca bu adımların tekrar edilmesi.



**Beklenen Sonuç ve Analitik Değerlendirme**

Amaç fonksiyonu incelendiğinde,

$$y = 5x_1 + 2(100-x_1) - x_1(100-x_1)$$$$y = x_1^2 - 97x_1 + 200$$
Bu parabolik fonksiyon, kısıtlı bölge ($30 \le x_1 \le 100$) içinde en büyük değerini sınır noktalarında alır.

- x_1 = 30 için: y = 5(30) + 2(70) - 30(70) = -1810
- x_1 = 100 için: y = 5(100) + 2(0) - 100(0) = 500

Teorik Olarak En Uygun Çözüm:

- x1 = 100
- x2 = 0
- y = 500


**Sonuç**

Genetik Algoritma, yüksek ceza katsayıları ve yeterli jenerasyon sayısı sayesinde kısıtlı bölgeyi başarıyla keşfetmiş ve elde edilen sonuçların bu sınır çözüme ($\mathbf{x_1 \approx 100, x_2 \approx 0, y \approx 500}$) yakınsadığı gözlemlenmiştir.Bu durum, algoritmanın problemi doğru şekilde modellediğini ve verilen kısıtlar altında global maksimum çözüme başarıyla ulaştığını göstermektedir. Elde edilen sonuçlar, matematiksel analiz ile tutarlı olup Genetik Algoritmanın bu problem için etkili bir optimizasyon yöntemi olduğunu doğrulamaktadır.

	​


```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. PROBLEM VE GA AYARLARI (SENARYO 2) ---
# Amaç: y = 5x1 + 2x2 - x1x2 (Maksimize edilecek)
# Kısıtlar: x1 + x2 = 100 ve x1 >= 30

POPULASYON_BOYUTU = 100 # Popülasyon büyüklüğü (Daha fazla çeşitlilik için artırıldı)
NESIL_SAYISI = 200 # İterasyon sayısı (Kısıtları oturtması için artırıldı)
MUTASYON_ORANI = 0.1 # Mutasyon oranı (Yerel minimumdan kaçış için)
MUTASYON_ADIMI = 2.0 # Mutasyon adımı (Gende ne kadar değişim olacağı)
TURNUVA_BOYUTU = 5 # Seçilim baskısı

# Değişken Sınırları [x_min, x_max]
SINIRLAR = [0, 100]

# --- 2. FONKSİYONLAR ---

def populasyon_baslat(pop_boyutu):
    """
    Popülasyonu Başlatma: Her birey [x1, x2] genden oluşur.
    Rastgele başlangıç değerleri üretilir.
    """
    # [Popülasyon Boyutu, 2] boyutunda matris
    return np.random.uniform(SINIRLAR[0], SINIRLAR[1], (pop_boyutu, 2))

def uygunluk_fonksiyonu(birey):
    """
    Amaç fonksiyonu ve Ceza (Penalty) Yöntemi.
    """
    x1 = birey[0]
    x2 = birey[1]

    # 1. Amaç Fonksiyonu Değeri (Skor)
    skor = 5*x1 + 2*x2 - x1*x2

    ceza = 0

    # --- KISIT 1: x1 + x2 = 100 (Eşitlik Kısıtı İçin Ceza) ---
    # Toplamın 100'den ne kadar saptığını buluyoruz.
    toplam_hata = abs((x1 + x2) - 100)
    # Bu hatayı büyük bir katsayı ile çarpıp ceza olarak ekliyoruz.
    ceza += toplam_hata * 1000

    # --- KISIT 2: x1 >= 30 (Eşitsizlik Kısıtı İçin Ceza) ---
    if x1 < 30:
        ihlal_miktari = 30 - x1
        ceza += ihlal_miktari * 1000 # İhlal miktarı kadar ceza

    # --- SINIR KISITLARI (0-100 Arası) ---
    if not (SINIRLAR[0] <= x1 <= SINIRLAR[1]) or not (SINIRLAR[0] <= x2 <= SINIRLAR[1]):
        ceza += 10000

    # Maksimizasyon problemi olduğu için Fitness = Skor - Ceza
    return skor - ceza

def secilim(populasyon, uygunluk_skorlari):
    """
    Turnuva Seçimi: Rastgele k birey seç, en iyisini al.
    """
    secilen_indeksler = np.random.randint(0, len(populasyon), TURNUVA_BOYUTU)
    en_iyi_idx = secilen_indeksler[0]
    for idx in secilen_indeksler:
        if uygunluk_skorlari[idx] > uygunluk_skorlari[en_iyi_idx]:
            en_iyi_idx = idx
    return populasyon[en_iyi_idx].copy()

def caprazlama(ebeveyn1, ebeveyn2):
    """
    Aritmetik Çaprazlama: Ebeveynlerin ağırlıklı ortalaması.
    """
    alfa = np.random.rand() # 0 ile 1 arasında rastgele ağırlık
    cocuk1 = alfa * ebeveyn1 + (1 - alfa) * ebeveyn2
    cocuk2 = (1 - alfa) * ebeveyn1 + alfa * ebeveyn2
    return cocuk1, cocuk2

def mutasyon(birey):
    """
    Gaussian Mutasyon: Genin üzerine rastgele küçük bir değer ekler.
    """
    for i in range(len(birey)):
        if np.random.rand() < MUTASYON_ORANI:
            rasgele_degisim = np.random.uniform(-MUTASYON_ADIMI, MUTASYON_ADIMI)
            birey[i] += rasgele_degisim
            # Sınırların dışına çıkarsa düzelt (Clamping)
            birey[i] = np.clip(birey[i], SINIRLAR[0], SINIRLAR[1])
    return birey

# --- 3. ANA DÖNGÜ (MAIN LOOP) ---

def genetik_algoritmayi_calistir():
    # Popülasyonu başlat
    populasyon = populasyon_baslat(POPULASYON_BOYUTU)

    en_iyi_gecmis = [] # Grafiği çizmek için en iyi skorları tutacağız
    en_iyi_cozum = None
    global_en_iyi_uygunluk = -float('inf')

    print("Genetik Algoritma Çalışıyor... (Senaryo 2)")
    print("-" * 50)

    for nesil in range(NESIL_SAYISI):
        # Tüm bireylerin uygunluk değerlerini hesapla
        uygunluk_skorlari = np.array([uygunluk_fonksiyonu(birey) for birey in populasyon])

        # En iyi bireyi bul ve sakla
        mevcut_en_iyi_idx = np.argmax(uygunluk_skorlari)
        mevcut_en_iyi_uygunluk = uygunluk_skorlari[mevcut_en_iyi_idx]

        # Global en iyiyi güncelle
        if mevcut_en_iyi_uygunluk > global_en_iyi_uygunluk:
            global_en_iyi_uygunluk = mevcut_en_iyi_uygunluk
            en_iyi_cozum = populasyon[mevcut_en_iyi_idx].copy()

        en_iyi_gecmis.append(global_en_iyi_uygunluk)

        # Yeni nesil oluşturma
        yeni_populasyon = []

        # Elitizm: En iyi bireyi kaybetmemek için doğrudan yeni nesle aktar
        yeni_populasyon.append(en_iyi_cozum)

        while len(yeni_populasyon) < POPULASYON_BOYUTU:
            ebeveyn1 = secilim(populasyon, uygunluk_skorlari)
            ebeveyn2 = secilim(populasyon, uygunluk_skorlari)

            cocuk1, cocuk2 = caprazlama(ebeveyn1, ebeveyn2)

            cocuk1 = mutasyon(cocuk1)
            cocuk2 = mutasyon(cocuk2)

            yeni_populasyon.append(cocuk1)
            if len(yeni_populasyon) < POPULASYON_BOYUTU:
                yeni_populasyon.append(cocuk2)

        populasyon = np.array(yeni_populasyon)

        # Her 20 nesilde bir bilgi ver
        if nesil % 20 == 0:
            print(f"Nesil {nesil}: En İyi Fitness = {global_en_iyi_uygunluk:.4f} (x1={en_iyi_cozum[0]:.2f}, x2={en_iyi_cozum[1]:.2f})")

    return en_iyi_cozum, en_iyi_gecmis

# --- 4. SONUÇLAR VE GÖRSELLEŞTİRME ---

en_iyi_cozum, gecmis = genetik_algoritmayi_calistir()

# Amaç fonksiyonu değerini (cezasız) hesapla
nihai_skor = 5*en_iyi_cozum[0] + 2*en_iyi_cozum[1] - en_iyi_cozum[0]*en_iyi_cozum[1]
toplam_oran = en_iyi_cozum[0] + en_iyi_cozum[1]

print("-" * 50)
print("SONUÇ ANALİZİ")
print(f"Pigment A (x1) : {en_iyi_cozum[0]:.4f} % (Kısıt: >=30)")
print(f"Pigment B (x2) : {en_iyi_cozum[1]:.4f} %")
print(f"Toplam Oran : {toplam_oran:.4f} % (Hedef: 100)")
print(f"Amaç Skoru (y) : {nihai_skor:.4f} (Maksimize Edilen Değer)")
print("-" * 50)

# Grafik Çizimi (Yakınsama Görselleştirmesi)
plt.figure(figsize=(10, 6))
plt.plot(gecmis, label='En İyi Fitness Değeri (Global)', color='navy', linewidth=2)
plt.title('Genetik Algoritma Yakınsama Grafiği (Senaryo 2)')
plt.xlabel('İterasyon (Nesil)')
plt.ylabel('Fitness (Uygunluk) Değeri')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
